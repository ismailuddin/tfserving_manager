```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import uuid
import os
import shutil
```

```python
from keras.applications.resnet50 import ResNet50
from keras.optimizers import Adam
from keras.layers import Input
```

```python
shutil.rmtree(os.path.join("..", "models", "src"))
os.mkdir(os.path.join("..", "models", "src"))
```

Models are saved in a source (`src`) directory, as these will need to be processed before they can be used by TensorFlow Serving

```python
for _ in range(3):
    MODEL_ID = str(uuid.uuid4())[:5].upper()
    model = ResNet50(weights="imagenet", input_tensor=Input((224, 224, 3)))
    model.compile(optimizer=Adam(), loss="categorical_crossentropy")
    model.save(os.path.join("..", "models", "src", f"model_{MODEL_ID}.h5"))
```
